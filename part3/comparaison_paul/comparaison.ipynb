# Testing

```python
# Pour récupérer les configs des environnements (pour la compatibilité)
env = gym.make("highway-fast-v0", render_mode='rgb_array')
config_highway_observation = env.config['observation']
config_highway_action = env.config['action']

env = gym.make("intersection-v0", render_mode='rgb_array')
config_intersection_observation = env.config['observation']
```

```python
def run(model_name, env_name):
    model_path = "highway/" + model_name
    video_path = "videos/" + model_name + "__" + env_name

    if model_name[:3] == 'dqn':
        model = DQN.load(model_path)
    elif model_name[:3] == 'ppo':
        model = PPO.load(model_path)

    env = gym.make(env_name, render_mode='rgb_array')
    # Compatibilité
    if model_name[4:] == "highway-fast-v0":
        env.config['observation'] = config_highway_observation
        if env_name == "intersection-v0":
            env.config['action'] = config_highway_action
            env.config['action']['target_speeds'] = [0, 4.5, 9]
    elif model_name[4:] == "intersection-v0":
        env.config['observation'] = config_intersection_observation

    env = record_videos(env, video_path)
    for episode in trange(10, desc='Test episodes'):
        (obs, info), done = env.reset(), False
        while not done:
            action, _ = model.predict(obs, deterministic=True)
            obs, reward, done, truncated, info = env.step(action)
    env.close()
    show_videos(video_path)
```
